Ex. 3: Count Vectorizer (done in email spam)

Apply all this to Dataset

```python
def message_clean(message):
    test_removed=[char for char in message if char not in string.punctuation]
    test_removed=''.join(test_removed)
    test_removed=[word for word in test_removed.split() if word not in stopwords.words('english')]
    return test_removed
```

```python
yelp_clean=yelp_1_5['text'].apply(message_clean)
```

```python
print(yelp_clean[0]) #cleaned up review
```

```python
print(yelp_1_5['text'][0]) #original review
```

Now apply Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(analyzer= message_clean)
yelp_vectorizer=vectorizer.fit_transform(yelp_1_5['text'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(yelp_vectorizer.toarray())
```

```python
yelp_vectorizer.shape
```

Training Model with all Dataset

```python
from sklearn.naive_bayes import MultinomialNB

nb_classifier=MultinomialNB()
label=yelp_1_5['stars'].values
```

```python
yelp_1_5['stars'].values
```

```python
nb_classifier.fit(yelp_vectorizer, label)
```

```python
#testing_sample=['amazing food! highly recommend']
testing_sample=['shit food, made me sick']
testing_sample_vectorizer=vectorizer.transform(testing_sample)

test_predict=nb_classifier.predict(testing_sample_vectorizer)
test_predict
```

Additional Feature: Term Frequency- Inverse Document Frequency (TF-IDF)

```python
X=yelp_tfidf
y=label

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y , test_size=0.2)

from sklearn.naive_bayes import MultinomialNB

nb_classifier=MultinomialNB()
nb_classifier.fit(X_train, y_train)

y_predict=nb_classifier.predict(X_test)
y_predict

cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True)
```
